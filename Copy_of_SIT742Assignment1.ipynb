# Part 2. Text Analysis

## 2.1 extract high requency words in description

```python
import re
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.probability import *
from itertools import chain
#from tqdm import tqdm
import codecs
```

```python
with open('stopwords.txt') as f:
    stop_words = f.read().splitlines()
stop_words = set(stop_words)
print(stop_words)
```

```python
# write your code here
# define your tokenize

tokenizer = RegexpTokenizer(r"\w+(?:[-']\w+)?") 

sentence_list = []
for title in text['description']:
   sentence = [w for w in tokenizer.tokenize(title) if w not in stop_words]
   sentence_list.append(sentence)
tokenized_words = [word for word in sentence for sentence in sentence_list]   
```

```python
# find top common words with document frequencies > 5000
# you may use function FreqDist() and sort()

from itertools import takewhile
freq = FreqDist(tokenized_words)
data = freq.most_common()
top_common = dict(takewhile(lambda x: x[1] > 5000, data))
keylist = top_common.keys()

sorted_top_common = {}

for key in sorted(keylist):
  sorted_top_common[key] = top_common[key]
  
sorted_top_common
```

```python
# save your table to 'top_common_words.txt'  
with open('top_common_words.txt','w') as file:
  file.write(json.dumps(sorted_top_common))
```

## 2.2 Find key words for describing Shiraz using TF-IDF

```python
# select 'description' from 'variety' eqaul to  'Shiraz' 
shiraz_sentence = text[text['variety']=='Shiraz']['description']
shiraz_sentence
```

```python
# use TfidfVectorizer to calculate TF-IDF score
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(analyzer='word', stop_words = 'english')

x = tfidf.fit_transform(shiraz_sentence)
idf = tfidf.idf_
tfids_values = dict(zip(tfidf.get_feature_names(), idf))
 
```

```python
# find words with TF-IDF score >0.4 and sort them
words_score_04 = list((key,value.round(2)) for (key, value) in tfids_values.items() if value> 0.4)
words_score_04.sort()
words_score_04
```

```python
# save your table to 'key_Shiraz.txt'   
with open('key_Shiraz.txt','w') as file:
  file.write(json.dumps(words_score_04))
```
